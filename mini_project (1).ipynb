```python
from pulp import LpProblem, LpVariable, lpSum, LpMinimize, LpBinary

# Filter eligible deliveries
eligible_df = df_balanced[df_balanced['DeliveryDuration'] <= 5].reset_index(drop=True)

# Create binary decision variables
choices = [LpVariable(f"select_{i}", cat='Binary') for i in range(len(eligible_df))]

# Define the optimization problem
model = LpProblem("Minimize_Transport_Cost", LpMinimize)

# Objective: minimize total transport cost
model += lpSum([choices[i] * eligible_df.loc[i, 'TransportCost'] for i in range(len(eligible_df))])

# Constraint: select exactly 10 deliveries
model += lpSum(choices) == 10

# Solve
model.solve()
```

```python
selected = [i for i in range(len(eligible_df)) if choices[i].value() == 1]
optimized_df = eligible_df.loc[selected]
```

```python
import matplotlib.pyplot as plt

# Total cost before optimization
original_cost = df_balanced[df_balanced['DeliveryDuration'] <= 5]['TransportCost'].sum()

# Total cost after optimization
optimized_cost = optimized_df['TransportCost'].sum()

# Plot
plt.figure(figsize=(6, 4))
plt.bar(['Before Optimization', 'After Optimization'], [original_cost, optimized_cost], color=['gray', 'green'])
plt.title("Transport Cost Comparison")
plt.ylabel("Total Cost")
plt.tight_layout()
plt.show()
```
